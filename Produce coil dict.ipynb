```python
def sticking_function(startlist=[], endlist=[]):
    def f(x):
        y = False
        for s,e in zip(startlist, endlist):
            if s <= x <= e:
                y = True
        return y
    return np.vectorize(f)
```

```python
f = sticking_function([0,20,np.nan], [10,30,np.nan])
f = sticking_function([np.nan],[np.nan])
x = np.linspace(-10, 50, 61)
y = f(x)
```
